#Line Plot


####1.You are analyzing the **flights** dataset from Seaborn, which contains airline passenger traffic data over time. Your goal is to visualize the trend in the number of passengers over the years while distinguishing different months.

###Tasks:

*  Load the flights dataset from Seaborn.
*  Create a line plot to show how the number of passengers (passengers) changes over years (year).
*  Use the month column as the hue parameter to differentiate trends for each month.
*  Apply a custom color palette (e.g., "Set2" or "coolwarm") for better visualization.
*  Add markers to each point to highlight individual data points.
*  Set appropriate axis labels and a title for clarity.
*  Modify the line style (dashes) to differentiate months instead of colors.
*  Use a moving average or aggregation to smooth out the data for better readability.

```
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("flights")
sns.lineplot(x="year",y="passengers",data=data,hue="month",palette="Set2",marker="o",style="month",estimator="mean")
plt.title("Number of passengers changes over years",fontsize=15)
plt.grid(alpha=0.3)
plt.show()
```

####You are working with the **mpg** dataset from Seaborn, which contains information about different car models, their fuel efficiency, and other attributes. Your goal is to analyze the trend of horsepower over model years while considering different car origins and handling a large dataset efficiently.

###Tasks:

* Load the mpg dataset and inspect the data.
* Create a line plot to show how horsepower (horsepower) changes over the model year (model_year).
* Use the origin column as the hue parameter to differentiate cars from different regions.
* Set different line styles based on origin.
* Adjust line width (size) based on origin, ensuring clarity in visualization.
* Modify the error bands:
  1. Use a median estimator instead of the mean.
  2. Set a 95% confidence interval (ci=95) for error shading.
  3. Customize the error bands' transparency (alpha) in err_kws.
* Ensure the data is sorted before plotting for correct visualization.
* Apply a custom color palette (e.g., Set2 or coolwarm).
* Add axis labels, a title, and a legend to enhance readability.
* Use facetgrid or col parameter to create multiple subplots based on the number of cylinders (cylinders), showing a separate plot for each cylinder type.

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = sns.load_dataset("mpg")
sns.lineplot(x="model_year",y="horsepower",data=data,hue="origin",
             style="origin",size="origin",
             dashes=True,marker="o",
             estimator="median",errorbar=('ci',95),err_kws={'alpha':0.5},
             sort=True,palette="Set2"
             )
plt.title("Horsepower changes over the model year",fontsize=15)
plt.xlabel("Model_Year",fontsize=11)
plt.ylabel("Horsepower",fontsize=11)
plt.grid(alpha=0.3)
plt.show()
```
